```python
import pandas as pd
from pathlib import Path
Path('../data/hist_prices.db').touch()
```

```python
import sqlite3
conn = sqlite3.connect('../data/hist_prices.db')
c = conn.cursor()
```

```python
# load the data into a Pandas DataFrame
aave = pd.read_csv('../data/crypto_historical_prices/coin_Aave.csv')
binance_coin = pd.read_csv('../data/crypto_historical_prices/coin_BinanceCoin.csv')
bitcoin = pd.read_csv('../data/crypto_historical_prices/coin_Bitcoin.csv')
cardano = pd.read_csv('../data/crypto_historical_prices/coin_Cardano.csv')
chain_link = pd.read_csv('../data/crypto_historical_prices/coin_ChainLink.csv')
cosmos = pd.read_csv('../data/crypto_historical_prices/coin_Cosmos.csv')
cryptocom_coin = pd.read_csv('../data/crypto_historical_prices/coin_CryptocomCoin.csv')
dogecoin = pd.read_csv('../data/crypto_historical_prices/coin_Dogecoin.csv')
eos = pd.read_csv('../data/crypto_historical_prices/coin_EOS.csv')
ethereum = pd.read_csv('../data/crypto_historical_prices/coin_Ethereum.csv')
iota = pd.read_csv('../data/crypto_historical_prices/coin_Iota.csv')
litecoin = pd.read_csv('../data/crypto_historical_prices/coin_Litecoin.csv')
monero = pd.read_csv('../data/crypto_historical_prices/coin_Monero.csv')
nem = pd.read_csv('../data/crypto_historical_prices/coin_NEM.csv')
polkadot = pd.read_csv('../data/crypto_historical_prices/coin_Polkadot.csv')
solana = pd.read_csv('../data/crypto_historical_prices/coin_Solana.csv')
stellar = pd.read_csv('../data/crypto_historical_prices/coin_Stellar.csv')
tether = pd.read_csv('../data/crypto_historical_prices/coin_Tether.csv')
tron = pd.read_csv('../data/crypto_historical_prices/coin_Tron.csv')
uniswap = pd.read_csv('../data/crypto_historical_prices/coin_Uniswap.csv')
usd_coin = pd.read_csv('../data/crypto_historical_prices/coin_USDCoin.csv')
wrapped_bitcoin = pd.read_csv('../data/crypto_historical_prices/coin_WrappedBitcoin.csv')
xrp = pd.read_csv('../data/crypto_historical_prices/coin_XRP.csv')
```

```python
c.execute('''CREATE TABLE aave (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE binance_coin (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE bitcoin (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE cardano (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE chain_link (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE cosmos (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE cryptocom_coin (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE dogecoin (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE eos (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE ethereum (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE iota (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE litecoin (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE monero (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE nem (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE polkadot (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE solana (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE stellar (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE tether (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE tron (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE uniswap (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE usd_coin (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE wrapped_bitcoin (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
c.execute('''CREATE TABLE xrp (SNo, Name, Symbol, Date, High, Low, Open, Close, Volume, Marketcap )''')
```

```python
# write the data to a sqlite table
aave.to_sql('aave', conn, if_exists='append', index = False)
binance_coin.to_sql('binance_coin', conn, if_exists='append', index = False)
bitcoin.to_sql('bitcoin', conn, if_exists='append', index = False)
cardano.to_sql('cardano', conn, if_exists='append', index = False)
chain_link.to_sql('chain_link', conn, if_exists='append', index = False)
cosmos.to_sql('cosmos', conn, if_exists='append', index = False)
cryptocom_coin.to_sql('cryptocom_coin', conn, if_exists='append', index = False)
dogecoin.to_sql('dogecoin', conn, if_exists='append', index = False)
eos.to_sql('eos', conn, if_exists='append', index = False)
ethereum.to_sql('ethereum', conn, if_exists='append', index = False)
iota.to_sql('iota', conn, if_exists='append', index = False)
litecoin.to_sql('litecoin', conn, if_exists='append', index = False)
monero.to_sql('monero', conn, if_exists='append', index = False)
nem.to_sql('nem', conn, if_exists='append', index = False)
polkadot.to_sql('polkadot', conn, if_exists='append', index = False)
solana.to_sql('solana', conn, if_exists='append', index = False)
stellar.to_sql('stellar', conn, if_exists='append', index = False)
tether.to_sql('tether', conn, if_exists='append', index = False)
tron.to_sql('tron', conn, if_exists='append', index = False)
uniswap.to_sql('uniswap', conn, if_exists='append', index = False)
usd_coin.to_sql('usd_coin', conn, if_exists='append', index = False)
wrapped_bitcoin.to_sql('wrapped_bitcoin', conn, if_exists='append', index = False)
xrp.to_sql('xrp', conn, if_exists='append', index = False)
```
